### Why Crime Rate?

A low crime rate is often one of the most important factors in determining a neighborhood's overwall quality of life. Safer communities generally experience more stable housing markets, higher resident satisfaction, and stronger social cohesion.

#### Dataset Used
- **Name:** Police Incident Blotter (ARCHIVED)
- **Source:** Western Pennsylvania Regional Data Center (WPRDC)
- **Link:** [Police Incident Blotter (WPRDC)](https://data.wprdc.org/dataset/police-incident-blotter)
- **Description:** This dataset contains incident-level crime reports filed by the City of Pittsburgh Police Department. It includes details such as type of offense, location, and neighborhood.

#### How I Used It
1. I focused on the **"INCIDENT NEIGHBORHOOD"** column to group crimes by neighborhood.
2. I cleaned the data by removing missing or unknown neighborhood entries.
3. I **counted the number of incidents** in each neighborhood
4. I **sorted the results from least to greatest**, since fewer reported crimes indicated higher safety.
5. I visualized the top 10 safest neighborhoods using:
   - **Vertical bar chart** for visualization
   - **Data table** for easy reference / clarity
  
#### Why It Matters
Crime directly affects how residents experience daily life. A neighborhood with fewer incidents is likely to feel safer and be more appealing for families, students, and working professinals alike. For this reason, I considered neighborhoods with the **least number of reported incidents** to be the "best" in terms of safety.

```python
import pandas as pd

# insert csv data
df = pd.read_csv("../kadair_walker/crime_data.csv")

# display data from cvs (defualt to first 10)
df.head(10)
```

```python
# display the columns of the dataset
df.columns
```

```python
# retrieve information from the "INCIDENTNEIGHBORHOOD" column
crime_counts = df["INCIDENTNEIGHBORHOOD"].value_counts()

# display 10 neighborhoods with most incidents reported
crime_counts.head(10)
```

```python
# display 10 neighborhoods with least incidents reported
crime_counts.tail(10)
```

```python
import matplotlib.pyplot as plt

# sorting crime_counts from least to greatest
safest = crime_counts.sort_values().head(10)

# plotting vertical bar graph
plt.figure(figsize=(10, 6))
plt.bar(safest.index, safest.values)

# titles and labels
plt.title("Top 10 Safest Neighborhoods in Pittsburgh")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Incidents")

# rotating x-axis labels to make them readable
plt.xticks(rotation = 45)

# display
plt.tight_layout()
plt.show()
```

```python
# remove or "clean" rows that do not display neighborhood names (NaN)
df_clean = df[df["INCIDENTNEIGHBORHOOD"].notna()]

# remove rows where the neighborhood is labeled as "Unknown"
df_clean = df_clean[df_clean["INCIDENTNEIGHBORHOOD"] != "Unknown"]

# counting number of incidents for each neighborhood
crime_counts = df_clean["INCIDENTNEIGHBORHOOD"].value_counts()
```

```python
import matplotlib.pyplot as plt

# sorting crime_counts from least to greatest
safest = crime_counts.sort_values().head(10)

# plotting vertical bar graph
plt.figure(figsize=(10, 6))
plt.bar(safest.index, safest.values)

# titles and labels
plt.title("Top 10 Safest Neighborhoods in Pittsburgh")
plt.xlabel("Neighborhood")
plt.ylabel("Number of Incidents")

# rotating x-axis labels to make them readable
plt.xticks(rotation = 45)

# display
plt.tight_layout()
plt.show()
```

```python
# creating a DataFrame from the 'safest' data
safest_df = pd.DataFrame({
    'Neighborhood': safest.index,
    'Number of Incidents': safest.values
})

# display
safest_df.reset_index(drop=True, inplace=True)
safest_df
```

This data ceased updating with the transition to a new records management system on 11/14/2023. Access to the updated data source will be provided in the future.

```python
import pandas as pd
df = pd.read_csv("../Rohan_Shrestha/Non_Traffic.csv")
df.head(20)
```

```python
# Filter to show only the NEIGHBORHOOD and COUNCIL_DISTRICT columns
filtered_df = df[['NEIGHBORHOOD', 'COUNCIL_DISTRICT']]

# Drop rows with missing council district BEFORE showing the top 10
filtered_df = filtered_df.dropna(subset=['COUNCIL_DISTRICT'])

# Now show the first 10 rows (all will have valid council district values)
# Get first 10 unique neighborhoods
unique_df = filtered_df.drop_duplicates(subset='NEIGHBORHOOD').head(10)
print(unique_df)
```

```python
# Drop rows with missing council district
filtered_df = filtered_df.dropna(subset=['COUNCIL_DISTRICT'])

# Get 10 unique neighborhoods directly (this avoids duplicates after selection)
plot_df = filtered_df.drop_duplicates(subset='NEIGHBORHOOD', keep='first').head(10)
```

```python
# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
barplot = sns.barplot(data=plot_df, x='NEIGHBORHOOD', y='COUNCIL_DISTRICT')

# Fix: Use enumerate so label positions match bars
for i, (neighborhood, row) in enumerate(plot_df.iterrows()):
    barplot.text(i, row['COUNCIL_DISTRICT'] + 0.1, str(int(row['COUNCIL_DISTRICT'])),
                 color='black', ha="center", fontsize=10)

plt.title('Council District of 10 Neighborhoods', fontsize=16)
plt.xlabel('Neighborhood', fontsize=12)
plt.ylabel('Council District', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

### Submetric 3 - Building Conditions (Alex)

### Why Building Conditions?

My submetric is average building condition. I chose this metric because it reflects the overall maintenance of the area. This allows for a fair comparison for all neighborhoods since it focuses on the general state of the buildings and not outliers

#### Dataset Used
- **Name:** Pittsburgh SNAP Census Data
- **Source:** Western Pennsylvania Regional Data Center (WPRDC)
- **Link:** [Pittsburgh SNAP Census Data](https://data.wprdc.org/dataset/pgh)
- **Description:** This dataset contains the  City of Pittsburgh SNAP census data from 2010. The data I will be focusing on is the building conditions from 2010.

#### How I Used It
1. I only focused on the column with the average percentage.
2. I cleaned the data by removing non-applicable entries.
3. I sorted the data to get the top 10 neighborhoods with the highest percentage.
4. I created a bar chart for visualization.

```python
import pandas as pd
df = pd.read_csv("../Alex_Nguyen/built-enviornment-conditions.csv")
condition_columns = [
    '% Good / Excellent Condition Buildings (2009)',
    '% Average Condition Buildings (2009)',
    '% Poor / Derelict Condition Buildings (2009)'
]
df_clean = df.dropna(subset=condition_columns, how='all')

df_clean
```

```python
df.columns
```

```python
neighborhood_percentages = df[["Neighborhood", "% Average Condition Buildings (2009)"]].dropna(subset=["% Average Condition Buildings (2009)"])

top_neighborhoods = neighborhood_percentages.sort_values(by="% Average Condition Buildings (2009)", ascending=True).tail(10)

top_neighborhoods
```

```python
import matplotlib.pyplot as plt


# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(top_neighborhoods["Neighborhood"], top_neighborhoods["% Average Condition Buildings (2009)"], color='skyblue')

# Add title and labels
plt.title("Top 10 Neighborhoods with Highest Average Building Conditions")
plt.xlabel("Neighborhoods")
plt.ylabel("Average Conditions Percentage")
plt.xticks(rotation=45)

# Adjust layout and display the plot
plt.show()
```
